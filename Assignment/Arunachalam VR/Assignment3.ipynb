**6.1.Identifying Outliers with Interquartile Range (IQR)**

```python
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
print((data > (Q3 + 1.5 * IQR))|(data < (Q1 - 1.5 * IQR)))
```

```python
import numpy as np
print(Q3, Q1)
print(IQR)

upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR
print(upper_bound)
print(lower_bound)

random_data = data[(data < upper_bound)|(data > lower_bound)]

print(random_data)
```

**6.2.Replacing Outliers with Median Values**

```python
a=data['Length'].quantile(0.50) 
b=data['Length'].quantile(0.95)
c=data['Diameter'].quantile(0.50) 
d=data['Diameter'].quantile(0.95)
e=data['Height'].quantile(0.50)
f=data['Height'].quantile(0.95) 
```

```python
data['Length'] = np.where(data['Length'] > b, a, data['Length'])
data['Diameter'] = np.where(data['Diameter'] > d, c, data['Diameter'])
data['Height'] = np.where(data['Height'] > f, e, data['Height'])

```

```python
a=data['Whole weight'].quantile(0.50) 
b=data['Whole weight'].quantile(0.95)
c=data['Shucked weight'].quantile(0.50) 
d=data['Shucked weight'].quantile(0.95)
e=data['Viscera weight'].quantile(0.50)
f=data['Viscera weight'].quantile(0.95) 
```

```python
data['Whole weight'] = np.where(data['Whole weight'] > b, a, data['Whole weight'])
data['Shucked weight'] = np.where(data['Shucked weight'] > d, c, data['Shucked weight'])
data['Viscera weight'] = np.where(data['Viscera weight'] > f, e, data['Viscera weight'])

```

```python
a=data['Shell weight'].quantile(0.50)
b=data['Shell weight'].quantile(0.95) 
c=data['Rings'].quantile(0.50)
d=data['Rings'].quantile(0.95) 
```

```python
data['Shell weight'] = np.where(data['Shell weight'] > b, a, data['Shell weight'])
data['Rings'] = np.where(data['Rings'] > d, c, data['Rings'])
```

**8.Splitting the dependent and independent columns**

```python
X = data.drop("Rings", axis=1)
X = X.values
```

```python
y = data["Rings"]
y = y.values
```

**9.Scale the independent variables**


```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler(feature_range =(0, 1)) 
new_x= min_max_scaler.fit_transform(X) 
print ("\n VALUES AFTER MIN MAX SCALING: \n\n", new_x) 
```

**10.Split the data into training and testing**

```python
from sklearn.model_selection import train_test_split 
X_train,X_test,y_train,y_test=train_test_split(new_x,y,test_size=0.20,random_state=10)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

**11.Build the model**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_train = knn.predict(X_train)
y_pred=knn.predict(X_test)
```

**12.Train the model**

```python
print("Training Accuracy : ",(metrics.accuracy_score(y_train, y_pred_train))*100)
```

**13.Test the model**

```python
print("Testing Accuracy : ",(metrics.accuracy_score(y_test, y_pred))*100)
```

**14.Measure the performance using metrics**

14.1.Confusion Matrix

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

14.2.Classification Report

```python
print(classification_report(y_test,y_pred))
```
